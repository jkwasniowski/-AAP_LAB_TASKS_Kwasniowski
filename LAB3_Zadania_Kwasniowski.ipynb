### Zadanie 1: Klasa Product + testy unittest

Stwórz klasę **Product** reprezentującą produkt w sklepie internetowym.

Klasa powinna zawierać:
- **Atrybuty:** `name` (str), `price` (float), `quantity` (int)
- **Metodę `add_stock(amount)`** — dodaje ilość do magazynu (rzuca `ValueError` przy ujemnej wartości)
- **Metodę `remove_stock(amount)`** — usuwa ilość z magazynu (rzuca `ValueError` jeśli za mało lub ujemna wartość)
- **Metodę `is_available()`** — zwraca `True` jeśli `quantity > 0`
- **Metodę `total_value()`** — zwraca `price * quantity`

**Kroki:**
1. Napisz klasę w pliku `product.py` (szkielet poniżej)
2. Napisz testy w pliku `test_product_unittest.py`
3. Użyj `setUp` do tworzenia instancji Product przed każdym testem
4. Przetestuj: poprawne działanie, wartości graniczne, obsługę błędów
5. Uruchom: `python -m unittest test_product_unittest -v`

Rozwiązanie klasy `Product` znajduje się w pliku `product.py`.

```python
from product import Product

produkt = Product("Laptop", 2999.99, 10)
print(produkt.name)
print(produkt.price)
print(produkt.quantity)
print(produkt.is_available())
print(produkt.total_value())
```

Testy `unittest` znajdują się w pliku `test_product_unittest.py`.

```python
import subprocess
import sys

result = subprocess.run(
    [sys.executable, "-m", "unittest", "test_product_unittest", "-v"],
    text=True,
    capture_output=True
)
print(result.stdout)
print(result.stderr)
assert result.returncode == 0
```

### Zadanie dodatkowe: Rozszerzenie klasy Product

Dodaj do klasy `Product` metodę **`apply_discount(percent)`**, która obniża cenę o podany procent.

**Wymagania:**
- `percent` musi być w zakresie 0-100 (w przeciwnym razie `ValueError`)
- Metoda zmienia atrybut `price`
- `apply_discount(50)` na produkcie z ceną 100.0 powinno ustawić cenę na 50.0

Napisz testy w pytest z `@pytest.mark.parametrize` dla różnych wartości procentowych:
- 0% (cena bez zmian)
- 50% (połowa ceny)
- 100% (cena = 0)
- Wartości ujemne (ValueError)
- Wartości > 100 (ValueError)

Testy metody `apply_discount` znajdują się w pliku `test_product_discount_pytest.py`.

```python
import subprocess
import sys

result = subprocess.run(
    [sys.executable, "-m", "pytest", "test_product_discount_pytest.py", "-v"],
    text=True,
    capture_output=True
)
print(result.stdout)
print(result.stderr)
assert result.returncode == 0
```
